# Join

Using two of the dataframes below, answer the questions :

```python
#import libraries here
import pandas as pd
```

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
                   
df2=pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

```python
df1
```

```python
df2
```

### Create a dataframe (df3) which consists both dataframe inner-joined by column "Key" (5 pts)

```python
#code goes here
df3= pd.merge(df1,df2,on='key')
df3
```

### Create a dataframe (df4) which consists of the result of df1 and df2 left-joined by column "Key" (10pts)

```python
#code goes here
df4= pd.merge(df1,df2,on='key',how='left')
df4
```

#### 1. Replace key "K2" into "K3" on df1
#### 2. Left join it with df2 
(15pts)

```python
#code goes here
df1_replaced = df1.replace("K2","K3")
df1_replaced

df1_replaced_left_df2 = pd.merge(df1_replaced,df2,on='key', how='left')
df1_replaced_left_df2
```

Show the average fare paid based on gender (10 pts)

```python
#code goes here
male = df.loc[df['Sex']=='male'].loc[:,'Fare'].mean()
female = df.loc[df['Sex']=='female'].loc[:,'Fare'].mean()

print('male average fare paid : ',male)
print('female average fare paid : ',female)
```

Show the total amount of passengers who survived and not survived based on gender (10pts)

```python
#code goes here
male_passengers = df.loc[df['Sex']=='male'].loc[:,'Survived'].value_counts()
female_passengers = df.loc[df['Sex']=='female'].loc[:,'Survived'].value_counts()
survivor = pd.DataFrame({
    '':['male','female'],
    'survived':[male_passengers[1],female_passengers[0]],
    'not survived':[male_passengers[0],female_passengers[1]]
})
survivor.style.hide_index()
```
